```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
df = ride_df.merge(city_df, how='left', on='city')
# Display the data table for preview
df.head()
```

```python
x = pd.DataFrame(df.groupby(["city", "type"])["ride_id"].agg(["count"]))
y = df.groupby(["city", "type"])["fare"].agg("mean")
```

```python
df_urban = df[df["type"].str.match("Urban")]
x_urban = df_urban.groupby("city")["ride_id"].agg("count")
y_urban = df_urban.groupby("city")["fare"].agg("mean")
s_urban = df_urban.groupby("city")["driver_count"].agg("mean")
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# Could have made a function but ran out of time, this is proof of concept that code works
df_urban = df[df["type"].str.match("Urban")]
x_urban = df_urban.groupby("city")["ride_id"].agg("count")
y_urban = df_urban.groupby("city")["fare"].agg("mean")
s_urban = df_urban.groupby("city")["driver_count"].agg("mean")

df_suburban = df[df["type"].str.match("Suburban")]
x_suburban = df_suburban.groupby("city")["ride_id"].agg("count")
y_suburban = df_suburban.groupby("city")["fare"].agg("mean")
s_suburban = df_suburban.groupby("city")["driver_count"].agg(["mean"])
            
df_rural = df[df["type"].str.match("Rural")]
x_rural = df_rural.groupby("city")["ride_id"].agg("count")
y_rural = df_rural.groupby("city")["fare"].agg("mean")
s_rural = df_rural.groupby("city")["driver_count"].agg(["mean"])

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, s=s_urban*12, c="LightCoral", alpha=0.8, edgecolors = "black", label = "Urban")
plt.scatter(x_suburban, y_suburban, s=s_suburban*12, c="Lightskyblue", alpha=0.8, edgecolors = "black", label = "Suburban")
plt.scatter(x_rural, y_rural, s=s_rural*12, c="gold", alpha=0.8, edgecolors = "black", label = "Rural")


# Create a legend
lgnd = plt.legend(numpoints=1, title = "City Types")
lgnd.legendHandles[0]._sizes=[10]
lgnd.legendHandles[1]._sizes=[10]
lgnd.legendHandles[2]._sizes=[10]


# Incorporate a text label regarding circle size
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Save Figure
plt.savefig('bubbleplot.png')
    
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Totals Fares by City Type
urban_count = df_urban["fare"].sum()
suburban_count = df_suburban["fare"].sum()
rural_count = df_rural["fare"].sum()
total_fares = urban_count + suburban_count + rural_count

# Calculate Type Percents
urban_per = urban_count / total_fares * 100
suburban_per = suburban_count / total_fares * 100
rural_per = rural_count / total_fares * 100


# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_per, rural_per, suburban_per]
explode = (0.1, 0, 0)
colors = ["LightCoral", "gold", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', pctdistance=0.6, shadow=True, labeldistance=1.1, startangle=-90, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False,)
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig('fares_by_city_type.png')
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Totals Fares by City Type
urban_ride_count = df_urban["ride_id"].count()
suburban_ride_count = df_suburban["ride_id"].count()
rural_ride_count = df_rural["ride_id"].count()
total_rides = urban_ride_count + suburban_ride_count + rural_ride_count

# Calculate Type Percents
urban_ride_per = urban_ride_count / total_rides * 100
suburban_ride_per = suburban_ride_count / total_rides * 100
rural_ride_per = rural_ride_count / total_rides * 100


# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_ride_per, rural_ride_per, suburban_ride_per]
explode = (0.1, 0, 0)
colors = ["LightCoral", "gold", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', pctdistance=0.6, shadow=True, labeldistance=1.1, startangle=-90, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False,)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig('rides_by_city_type.png')
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Totals Drivers by City Type
total_urban_drivers = df_urban.groupby("city")["driver_count"].max().sum()
total_suburban_drivers = df_suburban.groupby("city")["driver_count"].max().sum()
total_rural_drivers = df_rural.groupby("city")["driver_count"].max().sum()
total_drives =  total_urban_drivers + total_suburban_drivers +total_rural_drivers

# Calculate Type Percents
urban_drivers_per = total_urban_drivers / total_drives * 100
rural_drivers_per = total_rural_drivers / total_drives * 100
suburban_drivers_per = total_suburban_drivers / total_drives * 100

# Build Pie Chart
labels = ["Urban", "Rural", "Suburban"]
sizes = [urban_drivers_per, rural_drivers_per, suburban_drivers_per]
explode = (0.1, 0, 0)
colors = ["LightCoral", "gold", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', pctdistance=0.6, shadow=True, labeldistance=1.1, startangle=-90, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False,)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig('drivers_by_city_type.png')
```
